```python
#para generalizar, usan, tipo tensor
#=>
#z1 = a11X1+...a1nXn
#z2 = a21X1+...a2nXn
#zm = am1X1+...amnXn
#donde las dim son...
#x:(1,n+1)
#A:(n+1,m)#pues xÂ·A=z
#z:(1,m+1)

from matplotlib import pyplot as plt

import numpy as np
```

```python
N = 5 # N variables de entrada.
M = 3 # M variables objetivo.
P = 20 # P instancias
```

```python
x = np.random.uniform( -9, 9, (P,N+1))#los datos
x[:,-1]=1#lso datos mas un ultimo valor columna de 1,vias
A = np.random.normal(0,1,(N+1,M))#los parametros que uno busca, solo que los simulamos, para practicar esto
z = (np.dot( x, A))#los target, lo supervisado
```

```python
w = np.random.normal(0,0.1,(N+1,M))
```

```python
y = np.dot(x,w)
print(y)
```

```python
dist = z-y#el error de esos w
print(dist)
```

```python
delta_w = np.dot(x.T,dist)
print(delta_w)#los pesos entre dos capas de neuronas
```

```python
#error, np.sum(np.square(dist)) , o np.mean(np.saquare(dist))
#nos interesa iterar con tal de que el error sea menor a algo, y ademas, usar epocas=>una pasada por todos los datos
#es decir, usar un umbral de epocas aceptable
#learmimg rate, 0.01
```

```python
w = w + delta_w*0.01
print(w)
```

```python
error_linreg = np.sum(np.square(dist))
print(error_linreg)
```

```python
#hasta aca, son los pasos sucesivos, pra calcular el error, usemos estos en un ciclo
epoca = 0

errores_linreg = []
while(error_linreg > 0.001 and epoca < 100):
    
    y = np.dot(x,w)
    
    dist = z-y
    
    delta_w = np.dot(x.T,dist)
    
    w = w + delta_w*0.00001
    
    error_linreg = np.sum(np.square(dist))
    
    errores_linreg.append(error_linreg)
    
    epoca+=1
    
    if(epoca%20==0):
    
        print(error_linreg)
```

```python
#vale la pena graficar los errores
x_errores_linreg = np.arange(0,len(errores_linreg))

plt.plot(x_errores_linreg , errores_linreg)
```
