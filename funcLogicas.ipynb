```python
#que parneda simultanemanete el and y el or
import numpy as np
from matplotlib import pyplot as plt

N = 2 # N variables de entrada.
M = 2 # M variables objetivo.
P = 4 # P instancias.
#el input logico
logic_input = np.array( [ [ -1, -1 ],
[ -1, +1 ],
[ +1, -1 ],
[ +1, +1 ] ] )
#agregar vias a logic imput

logic_input_vias = np.concatenate((logic_input, np.ones((P,1))) , axis=1)

#col0 and , col1 or
logic_target = np.array( [ [ -1, -1 ],
[ -1, +1 ],
[ -1, +1 ],
[ +1, +1 ] ] )

#generemos los w al azar

w = np.random.normal(0,0.1,(N+1,M))

#el ciclo, para el perceptron...

#hasta aca, son los pasos sucesivos, pra calcular el error, usemos estos en un ciclo
epoca = 0
error_sign = 10000
errores_sign = []
while(error_sign > 0.001 and epoca < 1000):
    
    y = np.sign(np.dot(logic_input_vias,w))
    
    dist = logic_target-y
    
    delta_w = np.dot(logic_input_vias.T,dist)
    
    w = w + delta_w*0.001
    
    error_sign = np.sum(np.square(dist))
    
    errores_sign.append(error_sign)
    
    epoca+=1
    
    if(epoca%20==0):
    
        print(error_sign)

print(y)
```
