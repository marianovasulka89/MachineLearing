```python
# con Q = 0 Hebb, Q = 1 OJA ,Q = M OJA extendido, con Q = j Sanger

##algoritmo para no supervisado

#P filas , N col de entrada
#salida M filas
#definir un eta

#esto anda
import numpy as np

def NoSuperv(dataset , M = int(4) , eta = 0.001 , q = 4):

    """
    M es el num de neuronas de salida
    """

    N = dataset.shape[1]

    w = np.zeros((N , M))
#aca solo genero los pesos
    for i in range(0 , w.shape[0]):

        for j in range(0 , w.shape[1]):

            w[i][j] = np.random.uniform(-0.1 , 0.1)

    delta_w = np.zeros((N , M))

    y = np.zeros((1,M))

    print(f'dataset_dim: {dataset.shape}')

    for h in range(0,dataset.shape[0]):#hasta P

        x = dataset[h:h+1 , :]

        print(f'x_dim: {x.shape}')

        y[:] = np.dot(x , w)

        print(f'y_dim: {y.shape}')

        for j in range(0,M):

            for i in range(0,N):

                estimador_buena_entrada = 0

                for k in range(0,int(q)):

                    print(f'dim_y_k: {y[0,k].shape} dim_w_i_k: {w[i,k].shape} el producto: {y[0,k]} * {w[[i,k]]}')

                    estimador_buena_entrada += (y[0,k] * w[i,k])#elijo creer que pude arreglar esto

                print(f"EBE: {estimador_buena_entrada} y_i_k: {y[0,j]} w_i_k: {delta_w[i,k]}")

                delta_w[i,j] = ((x[0,i] - estimador_buena_entrada) * y[0,j])#aparentemente se rompe aca

        w += eta * delta_w

    return(w)
```
